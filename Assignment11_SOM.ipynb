1. Implement 1-D Self-organising Map (SOM) (own code) to cluster
the samples from Wine dataset and compare the performace
obtained with the in-built SOM library function.

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import random
import sys
import math
```

```python
wine = datasets.load_wine()
print(wine.feature_names)
data = pd.DataFrame(wine.data)

# normalize the dataset
max_val=max(data.max())
data =data/max_val

data.columns = wine.feature_names
data = np.array(data)
```

```python
def winner_cluster(x , W ,C):
  index = 0
  cost = np.linalg.norm(W[0] - x)
  for i in range(C):
    curr_cost = np.linalg.norm(W[i] - x)
    if(cost > curr_cost):
      curr_cost = cost
      index = i  
  return index 
```

```python
def Gaussian_Neighbourhood_funct( i, j):
  return math.exp(-((abs(i-j)**2) / 2))
```

```python
def SSE(mediod , X , cluster , k):
  sse_cost = 0
  for i in range(k):
    pattern = []
    for j in range(0, X.shape[0]):
      if(i == cluster[j]):
        pattern.append(j)
    for j in range(len(pattern)):
      sse_cost = sse_cost + np.linalg.norm(X[pattern[j]] - mediod[i])**2
  return sse_cost
```

```python
def SOM(X , C , epoch ):
  n = len(X[0])
  W = [[0]*n]*C
  
  for i in range(C):
    for j in range(n):
      W[i][j] = random.uniform(0 , 1)
    
  cluster = [0]*len(X)
  newCluster = [0]*len(X)
  for i in range(epoch):
    cluster = newCluster.copy()
    for j in range(len(X)):
      k = winner_cluster(X[j] , W , C)
      newCluster[j] = k
      # print(W)
      for l in range(C):
        h = Gaussian_Neighbourhood_funct(l , k)
        # here learning rate is 1
        W[l] = W[l] + (0.2)* h * X[j]*(X[j] - W[l])
    sse = SSE(W, X, cluster, C)
    print("Epoch: ", i+1, " SSE: ", sse)
    """
    if(cluster == newCluster):
      return cluster , W
    """
  return cluster , W
```
